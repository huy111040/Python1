```python
import requests
from bs4 import BeautifulSoup
# URL của trang web cần thu thập dữ liệu
url = 'https://xuanvinh.vn/collections/laptop-dell' 
# Gửi yêu cầu HTTP đến trang web
response = requests.get(url)
# Kiểm tra xem yêu cầu có thành công không
if response.status_code == 200:
  # Phân tích HTML
  soup = BeautifulSoup(response.text, 'html.parser')
  # Tìm tất cả các sản phẩm
  products = soup.find_all('div', class_='item_product_main')
  # Duyệt qua từng sản phẩm và thu thập thông tin
  for product in products:
    # Tên sản phẩm
    name = product.find('h3', class_='product-name').text.strip()
    # Đơn giá
    price = product.find('span', class_='price product-price').text.strip()
    # Giá gốc
    compare_price = product.find('span', class_='compare-price').text.strip()
    
    # In ra thông tin sản phẩm
    print(f'Tên sản phẩm: {name}')
    print(f'Giá sản phẩm: {price}')
    print(f'Giá gốc: {compare_price}')
    print('-' * 40)
else:
  print(f'Không thể truy cập trang web, mã trạng thái: {response.status_code}')
```

```python
import requests

url = 'https://randomuser.me/api/0.8/?results=10'

response = requests.get(url)

if response.status_code == 200:
    data = response.json()

    print('Gender list:')
    print('-' * 40)

    for i, item in enumerate(data["results"], start=1):
        gender = item["user"]["gender"]
        print(f'{i}. {gender}')
else:
    print(f'Không thể truy cập trang web, mã trạng thái: {response.status_code}')
```

```python
# Import thư viện pandas
import pandas as pd

# Đọc file CSV vào DataFrame
df = pd.read_csv("AB_NYC_2019.csv")

# Hiển thị thông tin tổng quát về dữ liệu (số dòng, cột, kiểu dữ liệu, non-null)
df.info()

# Hiển thị 5 dòng đầu tiên để xem dữ liệu mẫu
df.head()

# Đếm số dòng trùng lặp trong DataFrame
df.duplicated().sum()

# Đếm số lượng giá trị bị thiếu (NaN) trong từng cột
df.isna().sum()

# Tạo DataFrame mới bằng cách loại bỏ các dòng có giá trị NaN
df1 = df.dropna()

# Xem lại thông tin sau khi loại bỏ các dòng có NaN
df1.info()

# Kiểm tra các giá trị bị thiếu (True/False) trong df1
df1.isna()

# Đếm số lượng giá trị NaN còn lại trong từng cột của df1
df1.isna().sum()

# Hiển thị lại thông tin tổng quát của df1
df1.info()

# Tạo DataFrame mới (df2) bằng cách điền giá trị thay thế cho các ô NaN
df2 = df.fillna({"last_review": 0, "reviews_per_month": 0})

# Kiểm tra lại thông tin của df2 sau khi điền giá trị
df2.info()

# Đếm lại số lượng giá trị NaN trong df2
df2.isna().sum()

# Chuyển cột 'last_review' sang kiểu dữ liệu datetime
df["last_review"] = pd.to_datetime(df["last_review"])

# Kiểm tra kiểu dữ liệu sau khi chuyển đổi
df.info()

# Xóa các cột không cần thiết khỏi DataFrame
df.drop(["name", "id", "host_name", "last_review"], axis=1, inplace=True)

# Kiểm tra lại cấu trúc dữ liệu sau khi xóa cột
df.info()

# Lấy danh sách các giá trị duy nhất trong cột 'room_type'
df.room_type.unique()

# Đếm số lần xuất hiện của từng loại phòng trong cột 'room_type'
df.room_type.value_counts()

# Danh sách các cột cần chuyển sang kiểu dữ liệu category
col_name = ["neighbourhood_group", "neighbourhood", "room_type"]

# Chuyển 3 cột này sang kiểu category để tiết kiệm bộ nhớ và tối ưu xử lý
df[col_name] = df[col_name].astype("category")
```
